**Q.1.** Given an array **arr[ ]** of size **N** having elements, the task is to find the next greater element for each element of the array in order of their appearance in the array.Next greater element of an element in the array is the nearest element on the right which is greater than the current element.If there does not exist next greater of current element, then next greater element for current element is -1. For example, next greater of the last element is always -1.

**Example 1:**

```
Input:
N = 4, arr[] = [1 3 2 4]
Output:
3 4 4 -1
Explanation:
In the array, the next larger element
to 1 is 3 , 3 is 4 , 2 is 4 and for 4 ?
since it doesn't exist, it is -1.

```

**Example 2:**
    
```
Input:
N = 5, arr[] [6 8 0 1 3]
Output:
8 -1 1 3 -1
Explanation:
In the array, the next larger element to
6 is 8, for 8 there is no larger elements
hence it is -1, for 0 it is 1 , for 1 it
is 3 and then for 3 there is no larger
element on right and hence -1.
```

````python
# Function to find the next greater element for each element of the array
def nextGreater(arr, n):
  # Initialize a stack
  stack = []
  # Initialize an array to store the results
  result = [-1] * n
  # Loop through the array from right to left
  for i in range(n-1, -1, -1):
    # Pop the elements from the stack that are smaller than the current element
    while stack and stack[-1] <= arr[i]:
      stack.pop()
    # If the stack is not empty, the top element is the next greater element
    if stack:
      result[i] = stack[-1]
    # Push the current element to the stack
    stack.append(arr[i])
  # Return the result array
  return result

# Driver code
n = 4
arr = [1, 3, 2, 4]
print(nextGreater(arr, n))
````

**Q.2.** Given an array **a** of integers of length **n**, find the nearest smaller number for every element such that the smaller element is on left side.If no small element present on the left print -1.

**Example 1:**

```
Input: n = 3
a = {1, 6, 2}
Output: -1 1 1
Explaination: There is no number at the
left of 1. Smaller number than 6 and 2 is 1.
```

**Example 2:**
    
```
Input: n = 6
a = {1, 5, 0, 3, 4, 5}
Output: -1 1 -1 0 3 4
Explaination: Upto 3 it is easy to see
the smaller numbers. But for 4 the smaller
numbers are 1, 0 and 3. But among them 3
is closest. Similary for 5 it is 4.
```

````python
# Function to find the nearest smaller number for every element
def nearestSmaller(a, n):
  # Initialize a stack
  stack = []
  # Initialize an array to store the results
  result = []
  # Loop through the array from left to right
  for i in range(n):
    # Pop the elements from the stack that are greater than or equal to the current element
    while stack and stack[-1] >= a[i]:
      stack.pop()
    # If the stack is empty, there is no smaller element on the left
    if not stack:
      result.append(-1)
    # If the stack is not empty, the top element is the nearest smaller element on the left
    else:
      result.append(stack[-1])
    # Push the current element to the stack
    stack.append(a[i])
  # Return the result array
  return result

# Driver code
n = 3
a = [1, 6, 2]
print(nearestSmaller(a, n))
````

**Q.3.** Implement a Stack using two queues **q1** and **q2**.

**Example 1:**

```
Input:
push(2)
push(3)
pop()
push(4)
pop()
Output:3 4
Explanation:
push(2) the stack will be {2}
push(3) the stack will be {2 3}
pop()   poped element will be 3 the
        stack will be {2}
push(4) the stack will be {2 4}
pop()   poped element will be 4

```

**Example 2:**
    
```
Input:
push(2)
pop()
pop()
push(3)
Output:2 -1
```

````python
# Import the queue module
from queue import Queue

# Class to implement a stack using two queues
class Stack:
  # Constructor to initialize the queues
  def __init__(self):
    # Create two empty queues
    self.q1 = Queue()
    self.q2 = Queue()
  
  # Method to push an element to the stack
  def push(self, x):
    # Enqueue the element to the first queue
    self.q1.put(x)
  
  # Method to pop an element from the stack
  def pop(self):
    # If the first queue is empty, return -1
    if self.q1.empty():
      return -1
    # Dequeue all the elements from the first queue except the last one
    while self.q1.qsize() > 1:
      # Enqueue the dequeued element to the second queue
      self.q2.put(self.q1.get())
    # Dequeue and store the last element from the first queue
    x = self.q1.get()
    # Swap the names of the queues
    self.q1, self.q2 = self.q2, self.q1
    # Return the popped element
    return x

# Driver code
# Create a stack object
s = Stack()
# Push some elements to the stack
s.push(2)
s.push(3)
# Pop and print an element from the stack
print(s.pop())
# Push another element to the stack
s.push(4)
# Pop and print another element from the stack
print(s.pop())
````

Input:St = {3,2,1,7,6}

````python
# Define a stack class using a list
class Stack:
    # Constructor to initialize the list
    def __init__(self):
        self.items = []
  
    # Method to check if the stack is empty
    def empty(self):
        return len(self.items) == 0
  
    # Method to push an element to the stack
    def push(self, x):
        self.items.append(x)
  
    # Method to pop an element from the stack
    def pop(self):
        return self.items.pop()
  
    # Method to get the size of the stack
    def qsize(self):
        return len(self.items)

# Function to insert an element at the bottom of a stack
def insertAtBottom(st, x):
  # Base case: if the stack is empty, push the element
  if st.empty():
    st.push(x)
  # Recursive case: pop the top element, insert the element at the bottom, and push the popped element back
  else:
    y = st.pop()
    insertAtBottom(st, x)
    st.push(y)

# Function to reverse a stack using recursion
def reverseStack(st):
  # Base case: if the stack is not empty, pop the top element and reverse the remaining stack
  if not st.empty():
    x = st.pop()
    reverseStack(st)
    # Insert the popped element at the bottom of the reversed stack
    insertAtBottom(st, x)

# Driver code
# Create a stack object
st = Stack()
# Push some elements to the stack
st.push(3)
st.push(2)
st.push(1)
st.push(7)
st.push(6)
# Print the original stack
print(st.items)
# Reverse the stack
reverseStack(st)
# Print the reversed stack
print(st.items)
````

**Q.5.** You are given a string **S**, the task is to reverse the string using stack.

**Example 1:**
    
```
Input: S="GeeksforGeeks"
Output: skeeGrofskeeG
```

````python
# Define a stack class using a list
class Stack:
  # Constructor to initialize the list
  def __init__(self):
    self.items = []
  
  # Method to check if the stack is empty
  def empty(self):
    return len(self.items) == 0
  
  # Method to push an element to the stack
  def push(self, x):
    self.items.append(x)
  
  # Method to pop an element from the stack
  def pop(self):
    return self.items.pop()
  
  # Method to get the size of the stack
  def qsize(self):
    return len(self.items)

# Function to reverse a string using a stack
def reverseString(S):
  # Create a stack object
  st = Stack()
  # Loop through the string and push each character to the stack
  for c in S:
    st.push(c)
  # Initialize an empty string to store the reversed string
  rev = ""
  # Loop until the stack is empty and pop each character from the stack and append it to the reversed string
  while not st.empty():
    rev += st.pop()
  # Return the reversed string
  return rev

# Driver code
# Input string
S = "GeeksforGeeks"
# Print the original string
print(S)
# Reverse and print the string
print(reverseString(S))
````

**Q.6.** Given string **S** representing a postfix expression, the task is to evaluate the expression and find the final value. Operators will only include the basic arithmetic operators like ***, /, + and -**.

**Example 1:**

```
Input: S = "231*+9-"
Output: -4
Explanation:
After solving the given expression,
we have -4 as result.

```

**Example 2:**
    
```
Input: S = "123+*8-"
Output: -3
Explanation:
After solving the given postfix
expression, we have -3 as result.
```

````python
# Define a stack class using a list
class Stack:
    # Constructor to initialize the list
    def __init__(self):
        self.items = []
  
    # Method to check if the stack is empty
    def empty(self):
        return len(self.items) == 0
  
    # Method to push an element to the stack
    def push(self, x):
        self.items.append(x)
  
    # Method to pop an element from the stack
    def pop(self):
        return self.items.pop()
  
    # Method to get the size of the stack
    def qsize(self):
        return len(self.items)

# Function to evaluate a postfix expression
def evaluatePostfix(S):
    # Create a stack object
    st = Stack()
    # Loop through the string
    for c in S:
        # If the character is a digit, push it to the stack as an integer
        if c.isdigit():
            st.push(int(c))
        # If the character is an operator, pop two operands from the stack and perform the operation
        else:
            # Pop the right operand
            b = st.pop()
            # Pop the left operand
            a = st.pop()
            # Perform the operation and push the result to the stack
            if c == "+":
                st.push(a + b)
            elif c == "-":
                st.push(a - b)
            elif c == "*":
                st.push(a * b)
            elif c == "/":
                st.push(a / b)
    # Return the final value from the stack
    return st.pop()

# Driver code
# Input string
S = "231*+9-"
# Print the input string
print(S)
# Evaluate and print the postfix expression
print(evaluatePostfix(S))
````

**Q.7.** Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `MinStack` class:

- `MinStack()` initializes the stack object.
- `void push(int val)` pushes the element `val` onto the stack.
- `void pop()` removes the element on the top of the stack.
- `int top()` gets the top element of the stack.
- `int getMin()` retrieves the minimum element in the stack.

You must implement a solution with `O(1)` time complexity for each function.

**Example 1:**
    
```
Input
["MinStack","push","push","push","getMin","pop","top","getMin"]
[[],[-2],[0],[-3],[],[],[],[]]

Output
[null,null,null,null,-3,null,0,-2]

Explanation
MinStack minStack = new MinStack();
minStack.push(-2);
minStack.push(0);
minStack.push(-3);
minStack.getMin(); // return -3
minStack.pop();
minStack.top();    // return 0
minStack.getMin(); // return -2
```

````python
class MinStack:

    def __init__(self):
        # initialize a list to store the stack elements
        self.stack = []
        # initialize a variable to store the current minimum element
        self.min = None
        # initialize a list to store the output
        self.output = [None]

    def push(self, val: int) -> None:
        # if the stack is empty or the val is smaller than or equal to the current minimum
        if not self.stack or val <= self.min:
            # push the current minimum to the stack first
            self.stack.append(self.min)
            # update the current minimum to val
            self.min = val
        # push the val to the stack
        self.stack.append(val)
        # append None to the output
        self.output.append(None)

    def pop(self) -> None:
        # if the stack is not empty
        if self.stack:
            # pop the top element from the stack
            top = self.stack.pop()
            # if the top element is equal to the current minimum
            if top == self.min:
                # pop the next element from the stack, which is the previous minimum
                self.min = self.stack.pop()
        # append None to the output
        self.output.append(None)

    def top(self) -> int:
        # if the stack is not empty, return the top element of the stack
        if self.stack:
            # append the top element to the output
            self.output.append(self.stack[-1])
            return self.stack[-1]
        # otherwise, return None
        else:
            return None

    def getMin(self) -> int:
        # append the current minimum element to the output
        self.output.append(self.min)
        # return the current minimum element
        return self.min

    def getOutput(self) -> list:
        # return the output list
        return self.output

# test the code with the example input
minStack = MinStack()
minStack.push(-2)
minStack.push(0)
minStack.push(-3)
minStack.getMin() # returns -3
minStack.pop()
minStack.top()    # returns 0
minStack.getMin() # returns -2
print(minStack.getOutput()) 
````

**Q.8.** Given `n` non-negative integers representing an elevation map where the width of each bar is `1`, compute how much water it can trap after raining.

**Example 1:**
    

    
```
Input: height = [0,1,0,2,1,0,1,3,2,1,2,1]
Output: 6
Explanation: The above elevation map (black section) is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped.
```

**Example 2:**
    
```
Input: height = [4,2,0,3,2,5]
Output: 9
```

````python
def trap(height):
    # initialize left and right pointers
    left = 0
    right = len(height) - 1
    # initialize left and right max heights
    left_max = 0
    right_max = 0
    # initialize the result
    result = 0
    # loop until left and right pointers meet
    while left < right:
        # update the left max height
        left_max = max(left_max, height[left])
        # update the right max height
        right_max = max(right_max, height[right])
        # if left max is smaller than right max
        if left_max < right_max:
            # add the difference between left max and current height to the result
            result += left_max - height[left]
            # move the left pointer to the right
            left += 1
        # otherwise
        else:
            # add the difference between right max and current height to the result
            result += right_max - height[right]
            # move the right pointer to the left
            right -= 1
    # return the result
    return result

# test the code with the example input
height = [4,2,0,3,2,5]
print(trap(height))
````
